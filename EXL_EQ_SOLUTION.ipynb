```python
#parameters were tuned with the help of KAGGLE'S KERNEL
params = {
    'task': 'train',
    'objective': 'regression',
    'metric': {'l2', 'auc'},
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5
}
```

```python
names=list(store_map['Store_Name'])
r=range(len(names))
df=store_map.assign(showroom_no=r)
```

```python
df.head()
```

```python
df.drop(['Store_Name','Floor_name'],axis=1,inplace=True)
```

```python
y_train=df['showroom_no']
X_train=df[['Floor_Index','latitude','longitude']]
print(X_train.head())
print(y_train.head())
```

```python
dtrain = lgb.Dataset(X_train, y_train)
clf = lgb.train(params, dtrain,num_boost_round=100)
preds = clf.predict(X_train)
#print(r2_score(Y_val,preds)*200)
#dval = lgb.Dataset(X_val, Y_val)
```
